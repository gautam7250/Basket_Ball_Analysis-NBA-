```python
data = pd.read_csv(r'D:\Educational(B.E.)\Course\Data_Mining\sportsref_download.csv', parse_dates=['Date'])
```

```python
data1 = data.rename(columns={'Unnamed: 6':'Score Type', 'Visitor/Neutral':'Visitor Team', 'PTS':'VisitorPts','Home/Neutral':'Home Team','PTS.1':'HomePts','Unnamed: 7':'OT?','Notes':'Notes'})
```

```python
data2 = data1.drop(['Start (ET)','Attend.'], axis =1)
```

```python
data2.head()
```

```python
data2.shape
```

```python
data2["HomeWin"] = data2["VisitorPts"] < data2["HomePts"]
```

```python
y_true = data2["HomeWin"].values
```

```python
print("Home win Percentage: {0:.1f}%".format(100 * data2["HomeWin"].sum() / data2["HomeWin"].count()))
data2.head()
```

```python
data2["HomeLastWin"] = False
data2["VisitorLastWin"] = False
data2.head(10)
```

```python
from collections import defaultdict
```

```python
won_last = defaultdict()
```

```python
# Now compute the actual values for these
# Did the home and visitor teams win their last game?
from collections import defaultdict
won_last = defaultdict(int)

for index, row in data2.iterrows():  # Note that this is not efficient
    home_team = row["Home Team"]
    visitor_team = row["Visitor Team"]
    row["HomeLastWin"] = won_last[home_team]
    row["VisitorLastWin"] = won_last[visitor_team]
    data2.loc[index] = row    
    # Set current win
    won_last[home_team] = row["HomeWin"]
    won_last[visitor_team] = not row["HomeWin"]
data2.loc[20:25]
```

# Decision Making(Using Decision Tree)

<img src="training.png" width="800" height="400">

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state = 14)

from sklearn.model_selection import cross_val_score
```

```python
x_previouswins = data2[["HomeLastWin", "VisitorLastWin"]].values
scores = cross_val_score(clf, x_previouswins, y_true, scoring = 'accuracy')
```

```python
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

# Sport Outcome Predictions

```python
standings = pd.read_csv(r'D:\Educational(B.E.)\Course\Data_Mining\leagues_NBA_2013_standings_expanded-standings.csv', skiprows = [0])
```

```python
standings.head()
```

```python
# We can create a new feature -- HomeTeamRanksHigher\
data2["HomeTeamRanksHigher"] = 0
for index, row in data2.iterrows():
    home_team = row["Home Team"]
    visitor_team = row["Visitor Team"]
    if home_team == "New Orleans Pelicans":
        home_team = "New Orleans Hornets"
    elif visitor_team == "New Orleans Pelicans":
        visitor_team = "New Orleans Hornets"
    home_rank = standings[standings["Team"] == home_team]["Rk"].values[0]
    visitor_rank = standings[standings["Team"] == visitor_team]["Rk"].values[0]
    row["HomeTeamRanksHigher"] = int(home_rank > visitor_rank)
    data2.loc[index] = row
data2[:5]
```

```python
x_homehigher = data2[["HomeLastWin", "VisitorLastWin", "HomeTeamRanksHigher"]].values
```

```python
scores = cross_val_score(clf, x_homehigher, y_true, scoring='accuracy')
```

```python
print("Using wether the home team is ranked hhigher")
print("Accuracy: {0:.1f} %".format(np.mean(scores) * 100))
```

```python
last_match_winner = defaultdict(int)
data2["HomeTeamWonLast"] = 0
```

```python
for index, row in data2.iterrows():
    home_team = row["Home Team"]
    visitor_team = row["Visitor Team"]
    teams = tuple(sorted([home_team, visitor_team]))
    row["HomeTeamWonLast"] = 1 if last_match_winner[teams] == row["Home Team"] else 0
    data2.loc[index] = row
    winner = row["Home Team"] if row["HomeWin"] else row ["Visitor Team"]
    last_match_winner[teams] = winner

data2.loc[:5]
```

```python
X_home_higher =  data2[["HomeTeamRanksHigher", "HomeTeamWonLast"]].values
clf = DecisionTreeClassifier(random_state=14)
scores = cross_val_score(clf, X_home_higher, y_true, scoring='accuracy')
print("Using whether the home team is ranked higher")
print("Accuracy: {0:.1f}%".format(np.mean(scores) * 100))
```

```python
x_lastwinner = data2[["HomeTeamRanksHigher", "HomeTeamWonLast"]].values
clf = DecisionTreeClassifier(random_state = 14)
scores = cross_val_score(clf, x_lastwinner, y_true, scoring='accuracy')
print("Using wether the home team is ranked hhigher")
print("Accuracy: {0:.1f} %".format(np.mean(scores) * 100))
```
